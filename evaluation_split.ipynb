```python
import os
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("drive/MyDrive/9727/steam_reviews_sampled.csv")
```

```python
# Count number of reviews for each user
df_count = df['author.steamid'].value_counts().to_frame().reset_index().rename(columns={'author.steamid': 'count'})
```

```python
df_count.query("count >= 80")
```

```python
# Evaluation: only reserve users who made more than 80 reviews
evaluation_df = df.loc[df['author.steamid'].isin(df_count.query("count >= 80")['index'].values), :]
```

```python
# List of user who will be used for evaluation
user_list = evaluation_df['author.steamid'].unique()
```
